# **Heart Disease Prediction** System Using **SVM** (*support vector machine*)

## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Loading the dataset ( heart.csv )

```python
data = pd.read_csv("/content/heart (1).csv")
print(data.head())
```

```python
print(data.columns)
```

```python
print(data.info())
```

```python
print("Total rows : ", len(data.index))
print("Total columns : ", len(data.columns))
```

## Separating independent and dependent variables

```python
X = data.iloc[:, :10].values
y = data.iloc[:, -1:].values
```

```python
print(X[:5])
```

```python
print(y[:5])
```

## Spliting the dataset into training and testing 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train[:5])
```

```python
print(X_test[:5])
```

```python
print(y_train[:5])
```

```python
print(y_test[:5])
```

## Feature scaling using standardscaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_train[:5])
```

```python
print(X_test[:5])
```

## Training SVM model on training data

```python
from sklearn.svm import SVC
model = SVC(kernel="rbf")
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
print(y_predict[:5])
```

## Calculating accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_predict)
accuracy = accuracy_score(y_test, y_predict)*100
```

## Confusion matrix for better understanding

```python
print("confusion_matrix : \n", cm)
```

```python
print("Accuracy of the model : ", round(accuracy, 4), "%")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model_Randomforest = RandomForestClassifier(n_estimators=10, criterion = 'entropy', random_state = 0)
model_Randomforest.fit(X_train, y_train)
```

```python
y_predict_RF = model_Randomforest.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_predict_RF)
accuracy = accuracy_score(y_test, y_predict_RF)*100
```

```python
print("confusion_matrix : \n", cm)
```

```python
print("Accuracy of the RandomForest Model : ", round(accuracy, 4), "%")
```
